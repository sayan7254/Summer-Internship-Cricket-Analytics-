```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('M2.csv')
df.head()
```

```python
#Calculating the Given Parameters for the PowerPlay Overs
df.columns
```

```python
df=df[df['Over Number']<=6]
df.head()
```

```python
team=pd.DataFrame(df.groupby('Batting Team')['Batter Runs'].sum())
team
```

```python
team['Balls faced']=pd.DataFrame(df[df['Wides']==0].groupby('Batting Team')['Balls'].count())
team
```

```python
team['SR']=team['Batter Runs']/team['Balls faced']
team
```

```python
team['Economy']=team['Batter Runs']/6
team
```

```python
team['Sixes'] = pd.DataFrame(df[df['Runs'] == 6].groupby('Batting Team').size())
team
```

```python

team['Fours']=pd.DataFrame(df[df['Runs']==4].groupby('Batting Team').size())
team
```

```python
team['Boundaries']=team['Sixes']+team['Fours']
team
```

```python
df=df[df['Over Number']<=6]
```

```python
team['Dismissal Count']=df[df['Dismissal Type'].notna()].groupby('Batting Team').size()
team
```

```python
team['Batting Avg']=team['Batter Runs']/team['Dismissal Count']
team
```

```python
team['Total shot']=pd.DataFrame(df.groupby('Batting Team')['False Shot'].count())
team
```

```python
team['Fal Shot']=pd.DataFrame(df[df['False Shot']==1].groupby('Batting Team')['False Shot'].count())
team
```

```python
team['False Shot Percentages']=(team['Fal Shot']*100/team['Total shot'])
team
```

```python
team['Boundary Rate']=team['Boundaries']/team['Balls faced']
team
```

```python
team['Dismissal Rate']=team['Dismissal Count']/team['Balls faced']
team
```

```python
legal_balls = df[(df["Wides"] == 0) & (df["No Balls"] == 0)]
rotating_balls = legal_balls[legal_balls["Runs"].isin([1, 2, 3])]
total_balls = legal_balls.groupby("Batting Team").size()
rotating_counts = rotating_balls.groupby("Batting Team").size()
strike_rotation = (rotating_counts / total_balls * 100)
strike_rotation
```
